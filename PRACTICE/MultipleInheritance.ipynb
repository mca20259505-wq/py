```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def show_person(self):
        print("Name:", self.name, "Age:", self.age)


class Job:
    def __init__(self, job_title, salary):
        self.job_title = job_title
        self.salary = salary
    def show_job(self):
        print("Job Title:", self.job_title, "Salary:", self.salary)


class Manager(Person, Job):
    def __init__(self, name, age, job_title, salary, dept):
        Person.__init__(self, name, age)
        Job.__init__(self, job_title, salary)
        self.dept = dept

        
    def show_manager(self):
        self.show_person()
        self.show_job()
        print("Department:", self.dept)

mana = Manager("rehan", 20, "Manager", 2500000, "IT")
mana.show_manager()
```
