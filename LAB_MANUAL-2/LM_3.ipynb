#### 3.1: Pizzas: Think of at least three kind of your favorite pizza. Store these pizza names in a list, and  then use a for loop to print the name of each pizza.

#### 3.2: Animals: Think of at least three different animals that have a common characteristic. Store  the name of these animals in the list, and then use a for loop to,print out the name of each animal. 
#### • Modify your program to print a statement about each animal,such as A dog would make a great pet. 
#### • Add a line at the end of your program stating what these animals have in common. You could print a sentence such as Any of these animals would make a great pet! 

```python
print("Hujaifa")
print()
animals = ["dog","cat","rabbit"]
for animal in animals:
    print(animal)
print()
for animal in animals:
    if animal == "dog":
        print(f"A {animal} would make a great pet. They are loyal and playful!")
    elif animal == "cat":
        print(f"A {animal} would make a great pet. They are independent and cuddly!")
    elif animal == "rabbit":
        print(f"A {animal} would make a great pet. They are gentle and adorable!")
print()
print("Any of these animals would make a great pet!")
print("They are all friendly, lovable, and perfect companions!")
```

#### .5: Slices: Using one of the programs you wrote in this chapter, add several lines to the end of the program that do the following: 
#### • Print the message, the first three items in the list are :Then use a slice to print the first three items from that program's list. 
#### • Print the message, three items from the middle of the list are :Then use a slice to print three items from the middle of the list.
#### • Print the message, the last three items in the list are: Then use a slice to print the last three items in list. 

```python
print("Hujaifa")
pizzas = ["Margherita", "Pepperoni", "BBQ Chicken", "Hawaiian","Veggie Delight", "Four Cheese"]
print(pizzas)
print()
print(f"The first 3 pizzas : {pizzas[:3]}")
print(f"The last 3 pizzas : {pizzas[-3:]}")
middle_start = len(pizzas) // 2 - 1
middle_end = middle_start + 3
print(f"The mid 3 pizzas are : {pizzas[middle_start:middle_end]}")
```
