```python
print("Hujaifa")
lst = [x for x in range(20+1)]
fruits = ["apple","banana","mango","cherry","peach"]

print(lst)
print(fruits)
```

#### Concatenation of 2 lists

```python
print("Hujaifa")
print(lst + fruits)
```

####  length of liists

```python
print("Hujaifa")
print(len(lst))
print(len(fruits))
```

#### appending in list

```python
print("Hujaifa")
lst.append(100)
fruits.append("kiwi")
print(lst)
print(fruits)
```

#### popping 

```python
print("Hujaifa")
lst.pop()
fruits.pop()
print(lst)
print(fruits)
```

#### sorting


```python
print("Hujaifa")
lst.sort()
fruits.sort()
print(lst)
print(fruits)
```

#### reverse

```python
print("Hujaifa")
lst.reverse()
fruits.reverse()
print(lst)
print(fruits)
```

#### min()

```python
print("Hujaifa")
print(min(lst))
```
