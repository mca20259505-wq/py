#### 2.4 shrinking guests list: You just found out that your new dinner table won't arrive in time for the dinner and you have space for only two guests. 
#### • Start with your program from Exercise 2.3. Add a new line that prints a message saying that you can invite only two people for dinner. 
#### • Use pop() to remove guests from your list one at a time until only two names remain in your  list.Each time you pop a name from your list,print a message to that person letting them know you're sorry you can't invite them to dinner. 
#### • Print a message to each of the two people still on your list,letting them know they're still invited. 
#### • Use del to remove the last two names from your list,so you have an empty list.Print your list to make sure you actually have an empty list at the end of your program.

```python
print("Hujaifa")
guest_list = ["Marie Curie", "Albert Einstein", "Isaac Newton", 
              "Megji", "Nikola Tesla", "Grace Hopper"]

print("Original guest list:", guest_list)
print(f"Total guests: {len(guest_list)}\n")

print("Bad news! My new dinner table won't arrive in time.")
print("I can only invite 2 people for dinner.\n")


while len(guest_list) > 2:
    removed_guest = guest_list.pop()  
    print(f"Dear {removed_guest},")
    print("  I'm so sorry, but I no longer have space for you at the dinner.")
    print("  I hope we can meet another time!\n")


print("=== Good news for the remaining guests! ===\n")
new_guest_list = []
for guest in guest_list:
    print(f"Dear {guest},")
    print("  Great news! You're still invited to dinner!")
    print("  Looking forward to seeing you!\n")
    new_guest_list.append(guest)

del guest_list[0]  
del guest_list[0]  

print(f"remaining guests are : {new_guest_list}")
print("Final guest list:", guest_list)
print("List is now empty!" if not guest_list else "List still has guests!")
```

#### • Use sort() to print your list in alphabetical order without modifying the actual list.
#### • Show that your list is still in its original order by printing it. 
#### • Use sort() to print your list in reverse alphabetical order without changing the order of the  original list. 
#### • Show that your list is still in its original order by printing it again. 
#### • Use reverse() to change the order of youe list.Print the list to show that its order has changed. 
#### -Use reverse() to change the order of youe list again.Print the list to show its back to it’s original order. 
#### • Use sort() to change your list so it's stored in alphabetical order.Print the list to show that its order has been changed.
#### • Use sort() to change your list so it's stored in reverse alphabetical order. Print the list to show  that its order has been changed. 

```python
print("Hujaifa")
locations = ["shanghai","hell","heaven","tokyo"]
print(f"Original list : {locations}")
print("Sorted list :",sorted(locations))
print(f"Original list : {locations}")
print(f"reverse sorting : {sorted(locations, reverse=True)} ")
locations.reverse()
print(f"reverse : {locations}")
locations.reverse()
print(f"maintaining original order : {locations}")
locations.sort()
print(f"sorting : {locations}")
print(f"original order changed : {locations}")
locations.sort(reverse=True)
print(f"order changed again : {locations}")
```
